```python
df2=pd.read_csv('Testing1.csv');
```

```python
X=df2.iloc[:,:-1].values
Y=df2['prognosis'].values
```

```python
x=df.iloc[:,:-1].values
y=df['prognosis'].values
```

```python
x_train,x_test,y_train,y_test=train_test_split(x,y,test_size=0.15,random_state=2)
```

```python
sc_x=StandardScaler()
x_train=sc_x.fit_transform(x_train)
x_test=sc_x.transform(x_test)
```

```python
from sklearn.svm import SVC

s= SVC(kernel='linear')

s.fit(x_train, y_train)
acc=s.score(x_test,y_test)*100
print("Accuracy of svm: {:.2f}%".format(acc))
```

```python
#decision tree

accuracies = {}
#from sklearn.tree import DecisionTreeClassifier
dtc = DecisionTreeClassifier()
dtc.fit(x, y)
acc = dtc.score(X,Y)*100
accuracies['Decision Tree'] = acc
print("Decision Tree Test Accuracy {:.2f}%".format(acc))
```

```python
from sklearn.svm import SVC
svm = SVC(random_state = 0)
svm.fit(x, y)

acc = svm.score(X,Y)*100
accuracies['SVM'] = acc
print("Test Accuracy of SVM Algorithm: {:.2f}%".format(acc))
```

```python
from sklearn.neighbors import KNeighborsClassifier
knn = KNeighborsClassifier(n_neighbors = 8)  # n_neighbors means k
knn.fit(x_train, y_train)
prediction = knn.predict(x_test)

print("{} NN Score: {:.2f}%".format(2, knn.score(x_test,y_test)*100))
```

```python
from sklearn.ensemble import RandomForestClassifier
rf = RandomForestClassifier(n_estimators=2000,random_state=0)
rf.fit(x,y)

acc = rf.score(X,Y)*100
accuracies['Random Forest'] = acc
print("Random Forest Algorithm Accuracy Score : {:.2f}%".format(acc))
```
